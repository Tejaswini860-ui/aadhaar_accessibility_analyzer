```python
df["total_activity"] = (
    df["age_0_5"] + df["age_5_17"] + df["age_18_greater"]
)

print(df["total_activity"].describe())
```

```python
df["total_activity"] = (
    df["demo_age_5_17_y"].fillna(0) +
    df["demo_age_17__y"].fillna(0) +
    df["bio_age_5_17_y"].fillna(0) +
    df["bio_age_17__y"].fillna(0)
)

print(df["total_activity"].describe())
```

```python
threshold = df["total_activity"].median()

df["low_access"] = df["total_activity"].apply(
    lambda x: 1 if x < threshold else 0
)
```

```python
from sklearn.preprocessing import LabelEncoder

le_state = LabelEncoder()
le_district = LabelEncoder()

df["state_encoded"] = le_state.fit_transform(df["state"])
df["district_encoded"] = le_district.fit_transform(df["district"])
```

```python
features = ["state_encoded", "district_encoded"]

X = df[features]
Y = df["low_access"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

Y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, Y_pred))
```
