```python
import pandas as pd
df = pd.read_csv("../data/processed/final_merged_data.csv")
df = df.sample(n=50000, random_state=42)
df.head()
```

```python
df["total_activity"] = (
    df["age_0_5"] + df["age_5_17"] + df["age_18_greater"]
)

print(df["total_activity"].describe())
```

```python
df["total_activity"] = (
    df["demo_age_5_17_y"].fillna(0) +
    df["demo_age_17__y"].fillna(0) +
    df["bio_age_5_17_y"].fillna(0) +
    df["bio_age_17__y"].fillna(0)
)

print(df["total_activity"].describe())
```

```python
threshold = df["total_activity"].median()

df["low_access"] = df["total_activity"].apply(
    lambda x: 1 if x < threshold else 0
)
```

```python
from sklearn.preprocessing import LabelEncoder

le_state = LabelEncoder()
le_district = LabelEncoder()

df["state_encoded"] = le_state.fit_transform(df["state"])
df["district_encoded"] = le_district.fit_transform(df["district"])
```

```python
features = ["state_encoded", "district_encoded"]

X = df[features]
Y = df["low_access"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=50, max_depth=10)
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

Y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, Y_pred))
```

```python
low_regions = df[df["low_access"] == 1].copy()
low_regions[["state", "district"]].head()
```

```python
df["low_access"].value_counts()
```

```python
def generate_message(state, district):
    return f"Aadhaar enrollment/update is low in {district}, {state}. Visit nearest Aadhaar center."

low_regions["message"] = low_regions.apply(
    lambda row: generate_message(row["state"], row["district"]),
    axis=1
)

low_regions[["state", "district", "message"]].head()
```

```python
# Recreate X after sampling
X = df[["state_encoded", "district_encoded", "total_activity"]]

# -------------------------------
# CONFIDENCE SCORE
# -------------------------------
df["confidence"] = model.predict_proba(df[["state_encoded", "district_encoded"]])[:, 1]

# -------------------------------
# FUTURE PREDICTION
# -------------------------------
df["future_activity"] = df["total_activity"].shift(-1)
df["future_activity"] = df["future_activity"].fillna(df["future_activity"].median())

future_threshold = df["future_activity"].median()

df["future_risk"] = df["future_activity"].apply(
    lambda x: 1 if x <= future_threshold else 0
)

# -------------------------------
# RISK SCORE
# -------------------------------
df["risk_score"] = (
    df["low_access"] * 0.6 +
    (df["total_activity"] < threshold) * 0.4
)
```

```python
# -------------------------------
# FEATURE IMPORTANCE 
# -------------------------------
import pandas as pd

importance = model.feature_importances_

feature_importance_df = pd.DataFrame({
    "feature": ["state_encoded", "district_encoded"],
    "importance": importance
}).sort_values(by="importance", ascending=False)

print(feature_importance_df)
```
